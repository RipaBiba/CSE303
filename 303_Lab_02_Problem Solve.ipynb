```python
# Devisible by 08
def devisible_by_08(num):
    if num % 8 == 0:
        print(num)
        
# Use for the questions below:
nums = [i for i in range(1,1001)]
print("Divisible by 08: ")
for num in nums:
    devisible_by_08(num)
```

```python
# All numbers that have digit '6'
def in_num(num):
    if '6' in str(num):
        print(num)
        
# Use for the questions below:
nums = [i for i in range(1,1001)]
print("Have digit '6': ")
for num in nums:
    in_num(num)
```

## 03.

```python
# Have Spaaces
def check_space_in_string(string):
    count = 0
    
    for i in range(0, len(string)):
        if string[i] == " ":
            count += 1
         
    return count
 
string = "Practice Problems to Drill List Comprehension in Your Head."
print(f'The given String: "{string}"')
print(f'\nNumber of spaces in the given string: {check_space_in_string(string)}')
```

## 04.

```python
# Remove Vowels
def remove_vowels(string):
    newstring = string;
    print("\nRemoving vowels from the given string.....\n");
    vowels = ('a', 'e', 'i', 'o', 'u');
    for x in string.lower():
        if x in vowels:
            newstring = newstring.replace(x,"");
    print(f'New String after successfully removed all the vowels: "{newstring}"');

string = "Practice Problems to Drill List Comprehension in Your Head."
print(f'The given String: "{string}"')
remove_vowels(string)
```

## 05.

```python
# Less then 5 character
def less_than_5_character(string):
    short_string = []
    character = string.split(" ")
    for i in character:
        if len(i) < 5:
            short_string.append(i)
            
    return short_string
            
string = "Practice Problems to Drill List Comprehension in Your Head."
print(f'The given String: "{string}"')
print(f'\nThe words: {less_than_5_character(string)}')
```

## 06.

```python
# Count length of each word
def count_length(string):
    word = string.split()
    length = []
    for i in word:
        length.append(len(i))
    print(f'\nThe Words: {word}')
    print(f'\nThe Length of each Words: {length}')
    
string = "Practice Problems to Drill List Comprehension in Your Head."
print(f'The given String: "{string}"')
count_length(string)
```

```python
def find_max(nums):
    for num in nums:
        maximum = 0
        for i in range(1, 10):
            if (num % i) == 0:
                maximum = max(i, maximum)
        if(maximum != 0):
            print(f'For index: {num}, Maximum: {maximum}')
        
nums = [i for i in range(1,1001)]
find_max(nums)
```
